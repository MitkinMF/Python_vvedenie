Дана функция f(x) = -5*x**2 + 10*x - 150

1. Определить корни                                     

2. Найти интервалы, на которых функция возрастает       

3. Найти интервалы, на которых функция убывает          

4. Построить график                                     

5. Вычислить вершину                                    

6. Определить промежутки, на котором f > 0              

7. Определить промежутки, на котором f < 0              


```python
import math
from sympy import *
from sympy.plotting import plot as plt
# from sympy import Symbol
init_printing()

x = Symbol('x')
fx = -5*x**2 + 10*x - 150
```

1. Определить корни

```python

solve(fx) #мнимые корни т.к функция меньше 0
```

2. Найти интервалы, на которых функция возрастает

```python
solve(diff(fx)>0)
```

3. Найти интервалы, на которых функция убывает 

```python
solve(diff(fx)<0)
```

**4.1 Построение графика в промежутке х (-1,3)**

```python
p = plt(fx, (x,-1,3))
```

**4.2 Построение графика в промежутке х (-100,100)**

```python
p = plt(fx, (x,-100,100))
```


**5. Вершина**

```python
expr = solve(diff(fx))[0]
print(f'Координаты вершины : x={expr}, f(x)={fx.subs(x, expr)}')
```

6. Определить промежутки, на котором f > 0

```python
print(solve(fx>0))  #в данном случае функция всегда меньше 0
```

7. Определить промежутки, на котором f < 0

```python
print(solve(fx<0))   #функция меньше нуля везде
```
